```python
def getAvgPayoff(agent_list):
    average = 0
    for a in agent_list:
        average += a.getPayoff()
    return (average/len(agent_list))

def getAvgBias(agent_list):
    average = 0
    for a in agent_list:
        average += a.getBias()
    return (average/len(agent_list))
```

```python
bias_v_payoff = []
results_list = []
for i in range(len(model_list)):
    results_list.append([])
for i in range(len(model_list)):
    model = model_list[i]
    for f in model.faction_list:
    #     print(len(f.agent_list))
        ele = ["Faction " + str(f.getId()), getAvgPayoff(f.agent_list), getAvgBias(f.agent_list)]
        results_list[i].append(ele)
    ele = ["Group 2", getAvgPayoff(model.group_list[1]), getAvgBias(model.group_list[1])]
    results_list[i].append(ele)
for i in range(10):
    ele = ["Faction " + str(i), 0, 0]
    bias_v_payoff.append(ele)
bias_v_payoff.append(["Group 2", 0, 0])
avgPayoff = 0
avgBias = 0
for i in results_list:
    for ele in range(len(i)):
        bias_v_payoff[ele][1] += i[ele][1]
        bias_v_payoff[ele][2] += i[ele][2]
for i in bias_v_payoff:
    i[1] = i[1]/10
    i[2] = i[2]/10
```

```python


x_axis = []
y_axis = []
labels = []
for i in bias_v_payoff:
    y_axis.append(i[1])
    x_axis.append(i[2])
    labels.append(i[0])
fig, ax = plt.subplots(figsize =(12, 9))
plt.xlabel("Average Payoff", fontsize=15)
plt.ylabel("Average Bias", fontsize=15)
ax.scatter(x_axis, y_axis)

for i, txt in enumerate(labels):
    ax.annotate(txt, (x_axis[i], y_axis[i]), fontsize = 15)
```

```python
# PAYOFF AS A FUNCTION OF NUMBER OF STEPS
x_axis = []
y_axis1 = []
y_axis2 = []
for i in range(1, 1000):
    x_axis.append(100*i)
    y_axis1.append(0)
    y_axis2.append(0)

for model in model_list:
    l1 = model.payoff_list[0]
    l2 = model.payoff_list[1]
    for i in range(len(l1)):
        y_axis1[i] += l1[i]
        y_axis2[i] += l2[i]
for i in range(len(y_axis1)):
    y_axis1[i] = y_axis1[i]/10
    y_axis2[i] = y_axis2[i]/10

fig, ax = plt.subplots(figsize =(12, 9))
plt.xlabel("Number of iterations", fontsize=15)
plt.ylabel("Average Payoff of each agent", fontsize=15)
plt.plot(x_axis, y_axis1, color='r', label="Aligned (Group 1)")
plt.plot(x_axis, y_axis2, color='b', label="Unaligned (Group 2)")
plt.legend(fontsize=15)
```
